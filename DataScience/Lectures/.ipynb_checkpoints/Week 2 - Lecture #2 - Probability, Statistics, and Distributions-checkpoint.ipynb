## II. Deterministic vs. Probabilistic models:

### 1. Definitions:

__Deterministic Models__: prediction models that use known physical/scientific rules to predict outcomes.

__Probabilistic Models__: prediction models that use real/historical sample outcomes to predict unknown outcomes.

### 2. Comparison:

<img src="images/week_2/W2_Part_0_2_model_comparison.png" alt="Drawing" style="width: 500px;"/>

### 3. Summary:

__Deterministic models__ are useful where factors are limited/controlled, and the users are primarily interested in understanding the macro-relationships (i.e., in academia). Contrary, __probabilistic models__ are more useful in real-life predictions (i.e., in the industry).

### 2. Example

<img src="images/week_2/W2_Part_1_0_C_and_P_2.png" alt="Drawing" style="width: 500px;"/>

__iid__: independent and identically distributed

Answer by hand:
* Total possible outcomes = 5! = 5x4x3x2x1 = 120
* Total possible desirable outcomes = 2! x 3! = 2x1x3x2x1 = 12
* Probability of desirable outcomes (Total possible desirable outcomes / Total possible outcomes) = 12 / 120 = __1/10__

Answer with formula:
* C(5 2) = 5!/2!(5-2)! = __1/10__ OR
* C(5 3) = 5!/3!(5-3)! = __1/10__

Answer with code simulations:

```python
# itertools is helpful for combination/permutation:
from itertools import permutations, combinations

# declaring the 5 cars:
cars_set = {'Lyft_#1', 'Lyft_#2', 'Lyft_#3', 'Uber_#1', 'Uber_#2'}
# get all the ways in which the 5 cars can arrive:
permutation_cars_set = set(permutations(cars_set, 5))
# count the situation where the two Ubers arrive first:
counter = 0
# loop through all cases (120):
for loc_case in permutation_cars_set:
    # grab tuple of the first two cars:
    first_two_cars = loc_case[0:2]
    # check if the first two cars are Uber or Lyft:
    bool_first_two_cars =\
        [car_str.split('_')[0] =='Uber' for car_str in first_two_cars]
    # how many uber?
    uber_count = sum(bool_first_two_cars)
    # check the uber count for first two cars:
    if uber_count == 2:
        counter += 1
    else:
        continue
# print answer:
print('Probability where the first two cars are uber: {}'.format(
    float(counter)/len(permutation_cars_set)))
```

Answer with __conditional probability__:

* The probability of the first car being Uber is 2/5.
* If the first car is a Uber, the probability of the second car is Uber is 1/4.
* The probability of the first and second cars being Uber is: 2/5 x 1/4  = __1/10__

## II. Probability (Precursor of Statistics )

### 1. Bayes Rules

<img src="images/week_2/W2_Part_1_1_bayes_rules.png" alt="Drawing" style="width: 500px;"/>

__Keys__:
* A' = Not A
* B' = Not B
* Upside down 'U' = and
* P(A|B) = probability of A given B

#### Continue with previous Uber/Lyft problem:

The probability of second car is Uber given the first car is already Uber = __P (2nd = Uber | 1st = Uber)__

* __P (2nd = Uber | 1st = Uber)__ = P (1st and 2nd = Uber) / P(1st = Uber) = (1/10)/(2/5) = __1/4__

#### Bayes Rules (COVID example):

Terminology:
* __TP__: True Positive (Test True, Is True)
* __FN__: False Negative (Test False, Is True)
* __P__: Positive (Is True)
    * P = TP + FN
* __FP__: False Positive (Test True, Is False)
* __TN__: True Negative (Test False, Is False)
* __N__: Netagive (Is False)
    * N = FP + TN
* Note: __Pr = P__. Both denotes probability.
    
Given:
* Infection rate: __1% (0.01)__
    * aka. __Pr (are+)__
* True Positive Rate (Recall/Sensitivity): __93.8% (0.938)__
    * aka. __Pr (test+ | are+)__

    * aka. __Pr (test+ | are-)__

* __% of actual positive when tested positive__ (aka. you got tested positive and you are actually positive/infected)

### 2. Law of Total Probability

<img src="images/week_2/W2_Part_1_2_law_of_total_prob.png" alt="Drawing" style="width: 500px;"/>

The sub-probabilities have to equal the whole (__nothing more/less__)!

Example #1:
<img src="images/week_2/W2_Part_1_3_prob_ex_1.png" alt="Drawing" style="width: 500px;"/>

Answer:
* P (neither) = 1 - P (only genius) - P (only chocolate) - P (both)
    * P (only genius) = P (genius) - P (both) = 0.6 - 0.4 = 0.2
    * P (only chocolate) = P (chocolate) - P (both) = 0.7 - 0.4 = 0.3
* P (neither) = 1 - 0.2 - 0.3 - 0.4 = __0.1__
* __10%__ of the students are neither!

## III. Statistics (Precursor of Data Science)

### 1. Expectation

<img src="images/week_2/W2_Part_1_5_expectation.png" alt="Drawing" style="width: 500px;"/>

Example (Discrete): Expected Value of a Dice Roll (6 side dice)

<img src="images/week_2/E_dice.png" alt="Drawing" style="width: 500px;"/>

* __i__ = value of dice
* __1/6__ = probability of rolling each number
* __Expected Value__ = if we roll enough times, on average, we will roll this number.

Let's test it out with Python simulations using NumPy!

```python
# import NumPy:
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("fivethirtyeight")

# let's roll the dice 2000 times!
total_rolls = 2000
# list to record results:
results_lst = []
# list to record Expected Values up to a specific roll:
EV_lst = []
# simulations!:
for roll_number in range(1, total_rolls+1):
    loc_roll = np.random.randint(1, 7)
    results_lst.append(loc_roll)
    # calculate Expected Value thus far:
    loc_EV = np.mean(results_lst)
    # record EV:
    EV_lst.append(loc_EV)
# plot results:
plt.figure(figsize=(30, 10))
plt.title("Expectation Value of a Dice Roll")
plt.plot(EV_lst)
plt.ylabel('Expectation Value')
plt.xlabel('Number of dice rolls')
plt.show()
```

The __Variance__ shows __an average of how far away each dice roll is from the mean__. The value is squared here to make sure that the negative/positive distances from the mean don't cancel each other. In this case, we only care about the absolute distance and not the direction of the distance.

###  4. Covariance (we will re-visit this in detailed during the EDA session)

<img src="images/week_2/W2_Part_1_7_covariance.png" alt="Drawing" style="width: 500px;"/>

* Note: Correlation does not imply causation!
* Covariance to correlation = variance to standard deviation
* __Positive Covariance__: Indicates that two variables tend to move in the same direction.
* __Negative Covariance__: Reveals that two variables tend to move in inverse directions.

Example: Finance: index vs. company

Source:
https://corporatefinanceinstitute.com/resources/knowledge/finance/covariance/

```python
import pandas as pd

# input fake stock data:
data = np.array(
    [[2013, 2014, 2015, 2016, 2017],
    [1692, 1978, 1884, 2151, 2519],
    [68, 102, 110, 112, 154]]
).T
# create df:
fake_stock_df =\
    pd.DataFrame(
        data,
        columns=['Year', 'S&P500', 'X_company']
    )
# answer:
print('Covariance between S&P 500 and X_company is:')
print(fake_stock_df['S&P500'].cov(fake_stock_df['X_company']))
print('Positive covariance!')
```

Note:
* Formula above uses the sample covariance formula
* Direction (-/+) here doesn matter
* Single covariance value is a little different from the covariance matrix (will visit later)

# Week 1 Homework Review + Central Limit Theorem (CLT) 

```python
# read data in as df:
hw_1_df =\
    pd.read_csv(
        '/home/jupyter-admin-eric/coursework/DataScience'
        '/Data/500_Person_Gender_Height_Weight_Index.csv')
```

```python
# histogram of the population data:
plt.figure(figsize=(30, 10))
plt.title("Height Histogram (Total Population Size: 500)")
hw_1_df['Height'].hist(bins=100)
plt.ylabel('frequency (# of people)')
plt.xlabel('cm')
plt.show()
print('distribution not normal.')
```

```python
# different sample sizes:
sample_sizes = [1, 2, 30, 200, 300]
total_sample_run_per_size = 60
master_result_dict = dict()
master_means_per_size = dict()
# we will create 100 samples of the above respective sample sizes:
for loc_sample_size in sample_sizes:
    per_size_list = []
    per_size_means_list = []
    for loc_run in range(1, total_sample_run_per_size+1):
        loc_sample_arr = np.array(hw_1_df['Height'].sample(loc_sample_size))
        per_size_list.append(loc_sample_arr)
        per_size_means_list.append(np.mean(loc_sample_arr))
    master_result_dict[loc_sample_size] = per_size_list
    master_means_per_size[loc_sample_size] = per_size_means_list
```

```python
# plot out distribution of means per sample size:
for loc_size in sample_sizes:
    loc_means_list = master_means_per_size[loc_size]
    plt.figure(figsize=(30, 10))
    plt.title('Histogram of {} Sample Means for Sample Size {}'.format(
        str(total_sample_run_per_size), str(loc_size)))
    pd.Series(loc_means_list).hist()
    plt.ylabel('frequency samples')
    plt.xlabel('sample mean (cm)')
    plt.show()
```
