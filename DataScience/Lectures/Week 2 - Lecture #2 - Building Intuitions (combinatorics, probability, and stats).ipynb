<img src="images/week_2/W2_Part_1_0_C_and_P_2.png" alt="Drawing" style="width: 500px;"/>

__iid__: independent and identically distributed

### 1. Answer by hand:
* Total possible outcomes = 5! = 5x4x3x2x1 = 120
* Total possible desirable outcomes = 2! x 3! = 2x1x3x2x1 = 12
* Probability of desirable outcomes (Total possible desirable outcomes / Total possible outcomes) = 12 / 120 = __1/10__

### 2. Answer with formula:
* C(5 2) = 5!/2!(5-2)! = __1/10__ OR
* C(5 3) = 5!/3!(5-3)! = __1/10__

### 3. Answer with code simulations:

```python
# itertools is helpful for combination/permutation:
from itertools import permutations, combinations

# declaring the 5 cars:
cars_set = {'Lyft_#1', 'Lyft_#2', 'Lyft_#3', 'Uber_#1', 'Uber_#2'}
# get all the ways in which the 5 cars can arrive:
permutation_cars_set = set(permutations(cars_set, 5))
# count the situation where the two Ubers arrive first:
counter = 0
# loop through all cases (120):
for loc_case in permutation_cars_set:
    # grab tuple of the first two cars:
    first_two_cars = loc_case[0:2]
    # check if the first two cars are Uber or Lyft:
    bool_first_two_cars =\
        [car_str.split('_')[0] =='Uber' for car_str in first_two_cars]
    # how many uber?
    uber_count = sum(bool_first_two_cars)
    # check the uber count for first two cars:
    if uber_count == 2:
        counter += 1
    else:
        continue
# print answer:
print('Probability where the first two cars are uber: {}'.format(
    float(counter)/len(permutation_cars_set)))
```

### 4. Answer with __conditional probability__:

* The probability of the first car being Uber is 2/5.
* If the first car is a Uber, the probability of the second car is Uber is 1/4.
* The probability of the first and second cars being Uber is: 2/5 x 1/4  = __1/10__

### 2. Example (Discrete): Expected Value of a Dice Roll (6 side dice)

<img src="images/week_2/E_dice.png" alt="Drawing" style="width: 500px;"/>

#### Keys
* __i__ = value of dice
* __1/6__ = probability of rolling each number
* __Expected Value__ = if we roll enough times, the average rolled face values

#### Simulations (Let's test it out with Python simulations using NumPy!)

```python
# import NumPy:
import numpy as np
# import matplotlib:
import matplotlib.pyplot as plt
# style:
plt.style.use("fivethirtyeight")

# let's roll the dice 2000 times!
total_rolls = 2000
# list to record results:
results_lst = []
# list to record Expected Values up to a specific roll:
EV_lst = []
# simulations!:
for roll_number in range(1, total_rolls+1):
    loc_roll = np.random.randint(1, 7)
    results_lst.append(loc_roll)
    # calculate Expected Value thus far:
    loc_EV = np.mean(results_lst)
    # record EV:
    EV_lst.append(loc_EV)
# plot results:
plt.figure(figsize=(30, 10))
plt.title("Expectation Value of a Dice Roll")
plt.plot(EV_lst)
plt.ylabel('Expectation Value')
plt.xlabel('Nth Dice Rolls')
plt.show()
```

### 2. Example: Finance: Stock Index vs. Company Stock Value

### 2. Sampling (per different sizes)

```python
# different sample sizes:
sample_sizes = [1, 2, 30, 200, 300]
total_sample_run_per_size = 60
master_result_dict = dict()
master_means_per_size = dict()
# we will create 60 samples per above sample sizes:
for loc_sample_size in sample_sizes:
    per_size_list = []
    per_size_means_list = []
    for loc_run in range(1, total_sample_run_per_size+1):
        loc_sample_arr = np.array(hw_1_df['Height'].sample(loc_sample_size))
        per_size_list.append(loc_sample_arr)
        per_size_means_list.append(np.mean(loc_sample_arr))
    master_result_dict[loc_sample_size] = per_size_list
    master_means_per_size[loc_sample_size] = per_size_means_list
```

```python
# samples per size type:
master_result_dict
```

```python
# sample means per size type:
master_means_per_size
```

```python
# plot out distributions of sample means per sample size:
for loc_size in sample_sizes:
    loc_means_list = master_means_per_size[loc_size]
    plt.figure(figsize=(30, 10))
    plt.title('Histogram of {} Sample Means Where Each Has Sample Size {}'.format(
        str(total_sample_run_per_size), str(loc_size)))
    pd.Series(loc_means_list).hist()
    plt.ylabel('Frequency (# of Samples)')
    plt.xlabel('Sample Mean (cm)')
    plt.show()
```

### 3. Summary

When the __sample size is big enough__, the __sample mean__ of that sample is  from a __normally distributed sample means of the same sample size__ despite the original distribution of the data population.
